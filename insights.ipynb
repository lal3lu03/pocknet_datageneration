```python
# Analyze which columns have NaN values
print("=== BU48 Dataset (df_2) NaN Analysis ===")
nan_cols_bu48 = df_2.columns[df_2.isna().any()].tolist()
print(f"Columns with NaN values: {len(nan_cols_bu48)}")
print(f"Columns: {nan_cols_bu48}")

print("\n=== Chen11 Dataset (df) NaN Analysis ===")
nan_cols_chen11 = df.columns[df.isna().any()].tolist()
print(f"Columns with NaN values: {len(nan_cols_chen11)}")
print(f"Columns: {nan_cols_chen11}")

# Check if it's the same pattern
print(f"\nSame columns affected? {set(nan_cols_bu48) == set(nan_cols_chen11)}")
```

```python
# Examine the pattern - all NaN values seem to be from UNK residues
print("=== BU48 Dataset - NaN value analysis ===")
nan_rows_bu48 = df_2[df_2.isna().any(axis=1)]
print(f"Unique residue names in NaN rows: {nan_rows_bu48['residue_name'].unique()}")
print(f"Unique files with NaN values: {len(nan_rows_bu48['file_name'].unique())}")
print(f"Files: {nan_rows_bu48['file_name'].unique()}")

print("\n=== Chen11 Dataset - NaN value analysis ===")
nan_rows_chen11 = df[df.isna().any(axis=1)]
print(f"Unique residue names in NaN rows: {nan_rows_chen11['residue_name'].unique()}")
print(f"Unique files with NaN values: {len(nan_rows_chen11['file_name'].unique())}")
```

```python
# Check all UNK residues - they should all have NaN in chemical features
print("=== Checking all UNK residues in BU48 ===")
unk_rows_bu48 = df_2[df_2['residue_name'] == 'UNK']
unk_with_nan_bu48 = unk_rows_bu48[unk_rows_bu48.isna().any(axis=1)]
print(f"Total UNK residues: {len(unk_rows_bu48)}")
print(f"UNK residues with NaN: {len(unk_with_nan_bu48)}")
print(f"Percentage of UNK with NaN: {len(unk_with_nan_bu48)/len(unk_rows_bu48)*100:.2f}%")

print("\n=== Checking all UNK residues in Chen11 ===")
unk_rows_chen11 = df[df['residue_name'] == 'UNK']
unk_with_nan_chen11 = unk_rows_chen11[unk_rows_chen11.isna().any(axis=1)]
print(f"Total UNK residues: {len(unk_rows_chen11)}")
print(f"UNK residues with NaN: {len(unk_with_nan_chen11)}")
print(f"Percentage of UNK with NaN: {len(unk_with_nan_chen11)/len(unk_rows_chen11)*100:.2f}%")
```

```python
# Strategies for handling NaN values
print("=== NaN Value Handling Strategies ===")
print("\n1. REMOVE rows with NaN values:")
print(f"   BU48: Would remove {len(nan_rows_bu48)} rows ({len(nan_rows_bu48)/len(df_2)*100:.2f}% of data)")
print(f"   Chen11: Would remove {len(nan_rows_chen11)} rows ({len(nan_rows_chen11)/len(df)*100:.2f}% of data)")

print("\n2. IMPUTE with zeros (conservative approach):")
print("   - Chemical features for UNK residues set to 0")
print("   - Assumes unknown residues have no chemical properties")

print("\n3. IMPUTE with mean/median values:")
print("   - Use mean/median of known chemical features")
print("   - May introduce bias")

print("\n4. SEPARATE handling:")
print("   - Train separate models for UNK vs known residues")
print("   - More complex but potentially more accurate")

print("\nRECOMMENDATION:")
print("Since NaN values are concentrated in UNK residues (unknown/hetero atoms),")
print("and chemical features are undefined for these, the most reasonable approach")
print("is to IMPUTE with zeros, treating UNK residues as chemically neutral.")
```

```python
# Additional analysis: chem.atoms == 0 and residue mapping
print("\n=== Analysis of chem.atoms == 0 data points ===")
print("BU48 Dataset:")
chem_atoms_0_bu48 = df_2[df_2['chem.atoms'] == 0]
print(f"Rows with chem.atoms == 0: {len(chem_atoms_0_bu48)} ({len(chem_atoms_0_bu48)/len(df_2)*100:.2f}% of data)")
print(f"Unique residue names: {chem_atoms_0_bu48['residue_name'].unique()}")
print(f"Class distribution - Class 0: {len(chem_atoms_0_bu48[chem_atoms_0_bu48['class']==0])}, Class 1: {len(chem_atoms_0_bu48[chem_atoms_0_bu48['class']==1])}")

print("\nChen11 Dataset:")
chem_atoms_0_chen11 = df[df['chem.atoms'] == 0]
print(f"Rows with chem.atoms == 0: {len(chem_atoms_0_chen11)} ({len(chem_atoms_0_chen11)/len(df)*100:.2f}% of data)")
print(f"Unique residue names: {chem_atoms_0_chen11['residue_name'].unique()}")
print(f"Class distribution - Class 0: {len(chem_atoms_0_chen11[chem_atoms_0_chen11['class']==0])}, Class 1: {len(chem_atoms_0_chen11[chem_atoms_0_chen11['class']==1])}")
```

```python
# Analysis of non-standard residues
print("\n=== Non-standard residue analysis ===")
print("BU48 Dataset:")
for residue in ['MSE', 'HEM', 'UNK']:
    if residue in df_2['residue_name'].values:
        count = len(df_2[df_2['residue_name'] == residue])
        nan_count = len(df_2[df_2['residue_name'] == residue][df_2.isna().any(axis=1)])
        print(f"{residue}: {count} points, {nan_count} with NaN ({nan_count/count*100:.1f}% if count > 0 else 0)")
    else:
        print(f"{residue}: Not found")

print("\nChen11 Dataset:")
for residue in ['MSE', 'HEM', 'UNK']:
    if residue in df['residue_name'].values:
        count = len(df[df['residue_name'] == residue])
        nan_count = len(df[df['residue_name'] == residue][df.isna().any(axis=1)])
        print(f"{residue}: {count} points, {nan_count} with NaN ({nan_count/count*100:.1f}% if count > 0 else 0)")
    else:
        print(f"{residue}: Not found")
```

```python
# Improved strategies based on analysis
print("\n=== IMPROVED STRATEGIES ===")
print("\n1. REMOVE chem.atoms == 0 points:")
print("   - These points have no chemical neighborhood information")
print("   - Likely solvent-exposed or isolated points")
print("   - Would improve data quality by removing uninformative points")

print("\n2. RESIDUE MAPPING:")
print("   - MSE (selenomethionine) -> MET (methionine)")
print("   - Both have similar chemical properties, Se vs S")
print("   - This is biochemically justified")

print("\n3. HANDLE HEM (heme) specifically:")
print("   - Could map to custom properties or remove")
print("   - Heme has unique iron-containing porphyrin properties")

print("\n4. UNK residues:")
print("   - Set chemical features to 0 (neutral)")
print("   - Or remove if chem.atoms == 0")

print("\nRECOMMENDED APPROACH:")
print("1. Map MSE -> MET")
print("2. Remove all points where chem.atoms == 0")
print("3. Set remaining UNK/HEM chemical features to 0")
print("4. This should eliminate most NaN values while preserving data quality")
```
